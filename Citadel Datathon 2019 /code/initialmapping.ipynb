# Shape file

```python
import shapefile
```

```python
class World(object):
	"""docstring for Country"""
	def __init__(self):
		self.world = shapefile.Reader('world_admin_2008/world_admin_2008')
		shape = self.world.shapes()
		records = self.world.records()
		self._D = {}
		for i in range(len(shape)):
			cntry = records[i][6]
			self._D.setdefault(cntry, []).append(shape[i])
		self.ALL_COUNTRIES = self._D.keys()


	def __getitem__(self, key):
		if key not in self._D:	raise Exception(cntry + " not valid!!!")
		return self._D[key]

		
```

```python
world = World()
```

```python
colors = {"wind": "#2ecc71", "hydro":"#3498db","other": "#95a5a6","coal": "#e74c3c","petroleum": "#34495e"}
def to_color(x):
    if x in colors:
        return colors[x]
    else:
        return colors["other"]
```

```python
powerplants["colors"] = powerplants["primary_source"].apply(to_color)
```

```python
plt.scatter(range(5),range(5), c=["#2ecc71", "#3498db", "#95a5a6", "#e74c3c", "#34495e"])
```

```python
plt.figure(figsize=(15,10))
for shape in world['United States']:
    x = [p[0] for p in shape.points]
    y = [p[1] for p in shape.points]
    plt.fill(x,y, alpha = 0.1)
    
plt.scatter(powerplants["longitude"], powerplants["latitude"], s=powerplants["total_power"]/10, c=powerplants["colors"])
plt.scatter(range(5),range(5), c=["#2ecc71", "#3498db", "#95a5a6", "#e74c3c", "#34495e"], labels=)
plt.axis([-130, -60, 23,50])
```
